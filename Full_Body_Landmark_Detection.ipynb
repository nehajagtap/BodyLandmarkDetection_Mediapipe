```python
import cv2
import mediapipe as mp
```

```python
mp_drawing = mp.solutions.drawing_utils
mp_holistic = mp.solutions.holistic

# Get web camera feed

feed = cv2.VideoCapture(0)

try:
    while(feed.isOpened()):
        ret, frame = feed.read()
        cv2.imshow("Capture the movement", frame)

        # Exit on pressing 'q'
        if cv2.waitKey(10) & 0xFF == ord('q'):
            break

except KeyboardInterrupt:
    print("\nProgram interrupted by user. Closing Camera feed..")

finally:
    # Release resources
    feed.release()
    cv2.destroyAllWindows()
```

```python
# Detect facial keypoints
feed = cv2.VideoCapture(0)

try:
    # Initiate the holistic model
    with mp_holistic.Holistic(min_detection_confidence=0.5,min_tracking_confidence=0.5) as holistic:
        
        while(feed.isOpened()):
            ret, frame = feed.read()

            #Recolour the feed
            image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

            # Make detections
            detection = holistic.process(image)

            # Recolor imageback to BGR for rendering
            image =cv2.cvtColor(image, cv2.COLOR_RGB2BGR)

            # Draw Face Landmarks
            mp_drawing.draw_landmarks(image, detection.face_landmarks, mp_holistic.FACEMESH_TESSELATION)

            # Draw Right Hand Landmarks
            mp_drawing.draw_landmarks(image, detection.right_hand_landmarks, mp_holistic.HAND_CONNECTIONS)

            # Draw Left Hand Landmarks
            mp_drawing.draw_landmarks(image, detection.left_hand_landmarks, mp_holistic.HAND_CONNECTIONS)

            # Draw Pose Detections
            mp_drawing.draw_landmarks(image, detection.pose_landmarks, mp_holistic.POSE_CONNECTIONS)

            #Recolour the feed
            image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

            cv2.imshow("Capture the movement", image)

            if cv2.waitKey(10) & 0xff == ord('q'):
                break

except KeyboardInterrupt:
    print("\nProgram interrupted by user. Closing Camera feed..")

finally:
    # Release resources
    feed.release()
    cv2.destroyAllWindows()
```
